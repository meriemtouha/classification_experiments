Custom CNN

```python
import os, random, itertools
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, Subset, WeightedRandomSampler
from torchvision import datasets, transforms

from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import StratifiedShuffleSplit

# --- Reproducibility
SEED = 42
random.seed(SEED); np.random.seed(SEED)
torch.manual_seed(SEED); torch.cuda.manual_seed_all(SEED)
torch.backends.cudnn.deterministic = True; torch.backends.cudnn.benchmark = False



# --- Paths & Hyperparameters
DATA_DIR = "/content/drive/MyDrive/colon-data"
CHECKPOINT_PATH = "/content/best_colon_cnn.pt"

IMAGE_SIZE = 224   # resize from 768x768 -> 224x224
BATCH_SIZE = 32
EPOCHS = 20
LR = 1e-3
WEIGHT_DECAY = 1e-4
VAL_SPLIT, TEST_SPLIT = 0.15, 0.15
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", DEVICE)

# --- Transforms (resize step included!)
train_tfms = transforms.Compose([
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),   # resize large 768x768 to manageable size
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomVerticalFlip(p=0.1),
    transforms.RandomRotation(degrees=10),
    transforms.ColorJitter(brightness=0.15, contrast=0.15, saturation=0.15, hue=0.02),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
])
eval_tfms = transforms.Compose([
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),   # resize step for val/test
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
])

# --- Dataset & Splits
dataset_for_split = datasets.ImageFolder(root=DATA_DIR)
dataset_train_aug = datasets.ImageFolder(root=DATA_DIR, transform=train_tfms)
dataset_eval      = datasets.ImageFolder(root=DATA_DIR, transform=eval_tfms)

class_names = dataset_for_split.classes
targets = np.array([y for _, y in dataset_for_split.samples])
indices = np.arange(len(targets))

sss_test = StratifiedShuffleSplit(n_splits=1, test_size=TEST_SPLIT, random_state=SEED)
trainval_idx, test_idx = next(sss_test.split(indices, targets))
targets_trainval = targets[trainval_idx]
sss_val = StratifiedShuffleSplit(n_splits=1, test_size=VAL_SPLIT, random_state=SEED)
train_idx_rel, val_idx_rel = next(sss_val.split(trainval_idx, targets_trainval))
train_idx, val_idx = trainval_idx[train_idx_rel], trainval_idx[val_idx_rel]

train_ds, val_ds, test_ds = Subset(dataset_train_aug, train_idx), Subset(dataset_eval, val_idx), Subset(dataset_eval, test_idx)

# Weighted sampler for imbalance
train_targets = targets[train_idx]
class_sample_counts = np.bincount(train_targets, minlength=len(class_names)).astype(np.float32)
class_weights = 1.0 / np.maximum(class_sample_counts, 1.0)
sample_weights = class_weights[train_targets]
train_sampler = WeightedRandomSampler(weights=sample_weights, num_samples=len(sample_weights), replacement=True)

train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, sampler=train_sampler, num_workers=2, pin_memory=True)
val_loader   = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
test_loader  = DataLoader(test_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)

# --- Custom CNN
class SmallColonCNN(nn.Module):
    def __init__(self, num_classes=2):
        super().__init__()
        def conv_block(in_ch, out_ch):
            return nn.Sequential(
                nn.Conv2d(in_ch, out_ch, 3, padding=1),
                nn.BatchNorm2d(out_ch),
                nn.ReLU(inplace=True),
                nn.Conv2d(out_ch, out_ch, 3, padding=1),
                nn.BatchNorm2d(out_ch),
                nn.ReLU(inplace=True),
                nn.MaxPool2d(2)
            )
        self.features = nn.Sequential(
            conv_block(3, 32),
            conv_block(32, 64),
            conv_block(64, 128),
            conv_block(128, 256)
        )
        self.classifier = nn.Sequential(
            nn.Dropout(0.4),
            nn.Linear(256 * (IMAGE_SIZE // 16) * (IMAGE_SIZE // 16), 256),
            nn.ReLU(inplace=True),
            nn.Dropout(0.4),
            nn.Linear(256, num_classes)
        )
    def forward(self, x):
        x = self.features(x)
        x = torch.flatten(x, 1)
        return self.classifier(x)

model = SmallColonCNN(num_classes=2).to(DEVICE)

optimizer = torch.optim.Adam(model.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)
criterion = nn.CrossEntropyLoss()
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', factor=0.5, patience=2)


def accuracy_from_logits(logits, y):
    preds = torch.argmax(logits, dim=1)
    return (preds == y).float().mean().item()

# --- Training Loop
best_val_acc, history = 0.0, {"train_loss": [], "val_loss": [], "train_acc": [], "val_acc": []}
for epoch in range(1, EPOCHS+1):
    model.train()
    train_loss = train_acc = n_train = 0
    for X, y in tqdm(train_loader, desc=f"Epoch {epoch}/{EPOCHS} [Train]"):
        X, y = X.to(DEVICE), y.to(DEVICE)
        optimizer.zero_grad()
        logits = model(X)
        loss = criterion(logits, y)
        loss.backward(); optimizer.step()
        bs, acc = y.size(0), accuracy_from_logits(logits, y)
        train_loss += loss.item() * bs; train_acc += acc * bs; n_train += bs
    train_loss /= n_train; train_acc /= n_train

    model.eval(); val_loss = val_acc = n_val = 0
    with torch.no_grad():
        for X, y in val_loader:
            X, y = X.to(DEVICE), y.to(DEVICE)
            logits = model(X); loss = criterion(logits, y)
            bs, acc = y.size(0), accuracy_from_logits(logits, y)
            val_loss += loss.item() * bs; val_acc += acc * bs; n_val += bs
    val_loss /= n_val; val_acc /= n_val

    history["train_loss"].append(train_loss); history["val_loss"].append(val_loss)
    history["train_acc"].append(train_acc); history["val_acc"].append(val_acc)
    print(f"Epoch {epoch}: train_loss={train_loss:.4f} val_loss={val_loss:.4f} | train_acc={train_acc:.4f} val_acc={val_acc:.4f}")
    scheduler.step(val_acc)

    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), CHECKPOINT_PATH)
        print(f"  ✔ Saved new best model (val_acc={val_acc:.4f})")

print("Best val acc:", best_val_acc)

# --- Plots
epochs = range(1, len(history["train_loss"])+1)
plt.figure(); plt.plot(epochs, history["train_loss"], label="Train Loss")
plt.plot(epochs, history["val_loss"], label="Val Loss"); plt.legend(); plt.title("Loss vs Epochs"); plt.show()
plt.figure(); plt.plot(epochs, history["train_acc"], label="Train Acc")
plt.plot(epochs, history["val_acc"], label="Val Acc"); plt.legend(); plt.title("Accuracy vs Epochs"); plt.show()

# --- Load best & Test Evaluation
model.load_state_dict(torch.load(CHECKPOINT_PATH, map_location=DEVICE))
model.eval()
all_preds, all_labels = [], []
with torch.no_grad():
    for X, y in tqdm(test_loader, desc="Testing"):
        X = X.to(DEVICE)
        preds = torch.argmax(model(X), dim=1).cpu().numpy()
        all_preds.extend(preds); all_labels.extend(y.numpy())

cm = confusion_matrix(all_labels, all_preds)
print("Classification Report:\n", classification_report(all_labels, all_preds, target_names=class_names, digits=4))
print("Confusion Matrix:\n", cm)

def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion Matrix'):
    if normalize: cm = cm.astype('float') / cm.sum(axis=1, keepdims=True).clip(min=1)
    plt.figure(); plt.imshow(cm, interpolation='nearest'); plt.title(title); plt.colorbar()
    tick_marks = np.arange(len(classes)); plt.xticks(tick_marks, classes, rotation=45, ha='right'); plt.yticks(tick_marks, classes)
    fmt = '.2f' if normalize else 'd'; thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt), ha="center", color="white" if cm[i,j] > thresh else "black")
    plt.ylabel('True label'); plt.xlabel('Predicted label'); plt.tight_layout(); plt.show()

plot_confusion_matrix(cm, class_names, normalize=False, title='Confusion Matrix (Counts)')
plot_confusion_matrix(cm, class_names, normalize=True, title='Confusion Matrix (Normalized)')
```

Hybrid model (ResNet and Custom CNN)

```python


import os, random, itertools
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, Subset
from torchvision import datasets, transforms, models

from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import StratifiedShuffleSplit

# --- Reproducibility
SEED = 42
random.seed(SEED); np.random.seed(SEED)
torch.manual_seed(SEED); torch.cuda.manual_seed_all(SEED)
torch.backends.cudnn.deterministic = True; torch.backends.cudnn.benchmark = False


# --- Paths & Hyperparameters
DATA_DIR = "/content/drive/MyDrive/colon-data"
CHECKPOINT_PATH = "/content/best_hybrid.pt"

IMAGE_SIZE = 224
BATCH_SIZE = 16
EPOCHS = 20
LR = 1e-3
WEIGHT_DECAY = 1e-4
VAL_SPLIT, TEST_SPLIT = 0.15, 0.15
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", DEVICE)

# --- Transforms (no augmentation)
train_tfms = transforms.Compose([
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
])
eval_tfms = transforms.Compose([
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
])

# --- Dataset for splitting
dataset_for_split = datasets.ImageFolder(root=DATA_DIR)
targets = np.array([y for _, y in dataset_for_split.samples])
indices = np.arange(len(targets))
class_names = dataset_for_split.classes
print("Classes:", class_names)

# --- Train/Val/Test Split
sss_test = StratifiedShuffleSplit(n_splits=1, test_size=TEST_SPLIT, random_state=SEED)
trainval_idx, test_idx = next(sss_test.split(indices, targets))
targets_trainval = targets[trainval_idx]
sss_val = StratifiedShuffleSplit(n_splits=1, test_size=VAL_SPLIT, random_state=SEED)
train_idx_rel, val_idx_rel = next(sss_val.split(trainval_idx, targets_trainval))
train_idx, val_idx = trainval_idx[train_idx_rel], trainval_idx[val_idx_rel]

# ImageFolder twice with different transforms
train_ds_all = datasets.ImageFolder(root=DATA_DIR, transform=train_tfms)
eval_ds_all  = datasets.ImageFolder(root=DATA_DIR, transform=eval_tfms)

train_ds = Subset(train_ds_all, train_idx)
val_ds   = Subset(eval_ds_all, val_idx)
test_ds  = Subset(eval_ds_all, test_idx)

train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)
val_loader   = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
test_loader  = DataLoader(test_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)

# --- Hybrid Model: ResNet Backbone + Custom CNN Head
class HybridResNetCNN(nn.Module):
    def __init__(self, num_classes):
        super(HybridResNetCNN, self).__init__()
        backbone = models.resnet18(pretrained=True)
        self.feature_extractor = nn.Sequential(*list(backbone.children())[:-2])  # keep conv layers only

        for param in self.feature_extractor.parameters():
            param.requires_grad = False  # freeze ResNet backbone

        # Custom CNN head
        self.cnn_head = nn.Sequential(
            nn.Conv2d(512, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(256, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((1,1))
        )

        self.fc = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(64, num_classes)
        )

    def forward(self, x):
        x = self.feature_extractor(x)  # ResNet conv features
        x = self.cnn_head(x)           # custom CNN head
        x = self.fc(x)                 # classification
        return x

model = HybridResNetCNN(num_classes=len(class_names)).to(DEVICE)

# --- Optimizer & Loss
optimizer = torch.optim.Adam(model.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)
criterion = nn.CrossEntropyLoss()
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', factor=0.5, patience=2)

def accuracy_from_logits(logits, y):
    preds = torch.argmax(logits, dim=1)
    return (preds == y).float().mean().item()

# --- Training Loop
best_val_acc, history = 0.0, {"train_loss": [], "val_loss": [], "train_acc": [], "val_acc": []}
for epoch in range(1, EPOCHS+1):
    model.train()
    train_loss = train_acc = n_train = 0
    for X, y in tqdm(train_loader, desc=f"Epoch {epoch}/{EPOCHS} [Train]"):
        X, y = X.to(DEVICE), y.to(DEVICE)
        optimizer.zero_grad()
        logits = model(X)
        loss = criterion(logits, y)
        loss.backward(); optimizer.step()
        bs, acc = y.size(0), accuracy_from_logits(logits, y)
        train_loss += loss.item() * bs; train_acc += acc * bs; n_train += bs
    train_loss /= n_train; train_acc /= n_train

    model.eval(); val_loss = val_acc = n_val = 0
    with torch.no_grad():
        for X, y in val_loader:
            X, y = X.to(DEVICE), y.to(DEVICE)
            logits = model(X); loss = criterion(logits, y)
            bs, acc = y.size(0), accuracy_from_logits(logits, y)
            val_loss += loss.item() * bs; val_acc += acc * bs; n_val += bs
    val_loss /= n_val; val_acc /= n_val

    history["train_loss"].append(train_loss); history["val_loss"].append(val_loss)
    history["train_acc"].append(train_acc); history["val_acc"].append(val_acc)
    print(f"Epoch {epoch}: train_loss={train_loss:.4f} val_loss={val_loss:.4f} | train_acc={train_acc:.4f} val_acc={val_acc:.4f}")
    scheduler.step(val_acc)

    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), CHECKPOINT_PATH)
        print(f"  ✔ Saved new best model (val_acc={val_acc:.4f})")

print("Best val acc:", best_val_acc)

# --- Plots
epochs = range(1, len(history["train_loss"])+1)
plt.figure(); plt.plot(epochs, history["train_loss"], label="Train Loss")
plt.plot(epochs, history["val_loss"], label="Val Loss"); plt.legend(); plt.title("Loss vs Epochs"); plt.show()
plt.figure(); plt.plot(epochs, history["train_acc"], label="Train Acc")
plt.plot(epochs, history["val_acc"], label="Val Acc"); plt.legend(); plt.title("Accuracy vs Epochs"); plt.show()

# --- Load best & Test Evaluation
model.load_state_dict(torch.load(CHECKPOINT_PATH, map_location=DEVICE))
model.eval()
all_preds, all_labels = [], []
with torch.no_grad():
    for X, y in tqdm(test_loader, desc="Testing"):
        X = X.to(DEVICE)
        preds = torch.argmax(model(X), dim=1).cpu().numpy()
        all_preds.extend(preds); all_labels.extend(y.numpy())

cm = confusion_matrix(all_labels, all_preds)
print("Classification Report:\n", classification_report(all_labels, all_preds, target_names=class_names, digits=4))
print("Confusion Matrix:\n", cm)

def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion Matrix'):
    if normalize: cm = cm.astype('float') / cm.sum(axis=1, keepdims=True).clip(min=1)
    plt.figure(); plt.imshow(cm, interpolation='nearest'); plt.title(title); plt.colorbar()
    tick_marks = np.arange(len(classes)); plt.xticks(tick_marks, classes, rotation=45, ha='right'); plt.yticks(tick_marks, classes)
    fmt = '.2f' if normalize else 'd'; thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt), ha="center", color="white" if cm[i,j] > thresh else "black")
    plt.ylabel('True label'); plt.xlabel('Predicted label'); plt.tight_layout(); plt.show()

plot_confusion_matrix(cm, class_names, normalize=False, title='Confusion Matrix (Counts)')
plot_confusion_matrix(cm, class_names, normalize=True, title='Confusion Matrix (Normalized)')
```

```python

import os, random, itertools
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, Subset
from torchvision import datasets, transforms, models

from sklearn.metrics import confusion_matrix, classification_report
from sklearn.manifold import TSNE
import umap.umap_ as umap
from sklearn.model_selection import StratifiedShuffleSplit

import cv2
import matplotlib.cm as cm


SEED = 42
random.seed(SEED); np.random.seed(SEED)
torch.manual_seed(SEED); torch.cuda.manual_seed_all(SEED)
torch.backends.cudnn.deterministic = True; torch.backends.cudnn.benchmark = False


DATA_DIR = "/content/drive/MyDrive/colon-data"
CHECKPOINT_PATH = "/content/best_hybrid.pt"

IMAGE_SIZE = 224
BATCH_SIZE = 16
EPOCHS = 20
LR = 1e-3
WEIGHT_DECAY = 1e-4
VAL_SPLIT, TEST_SPLIT = 0.15, 0.15
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", DEVICE)


train_tfms = transforms.Compose([
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
])
eval_tfms = transforms.Compose([
    transforms.Resize((IMAGE_SIZE, IMAGE_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]),
])

# --- Dataset splitting
dataset_for_split = datasets.ImageFolder(root=DATA_DIR)
targets = np.array([y for _, y in dataset_for_split.samples])
indices = np.arange(len(targets))
class_names = dataset_for_split.classes
print("Classes:", class_names)

# --- Train/Val/Test Split
sss_test = StratifiedShuffleSplit(n_splits=1, test_size=TEST_SPLIT, random_state=SEED)
trainval_idx, test_idx = next(sss_test.split(indices, targets))
targets_trainval = targets[trainval_idx]
sss_val = StratifiedShuffleSplit(n_splits=1, test_size=VAL_SPLIT, random_state=SEED)
train_idx_rel, val_idx_rel = next(sss_val.split(trainval_idx, targets_trainval))
train_idx, val_idx = trainval_idx[train_idx_rel], trainval_idx[val_idx_rel]

# ImageFolder with transforms
train_ds_all = datasets.ImageFolder(root=DATA_DIR, transform=train_tfms)
eval_ds_all  = datasets.ImageFolder(root=DATA_DIR, transform=eval_tfms)

train_ds = Subset(train_ds_all, train_idx)
val_ds   = Subset(eval_ds_all, val_idx)
test_ds  = Subset(eval_ds_all, test_idx)

train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)
val_loader   = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
test_loader  = DataLoader(test_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)

# --- Hybrid Model: ResNet Backbone + Custom CNN Head
class HybridResNetCNN(nn.Module):
    def __init__(self, num_classes):
        super(HybridResNetCNN, self).__init__()
        backbone = models.resnet18(pretrained=True)
        # keep everything up to the last conv block (exclude avgpool and fc)
        self.feature_extractor = nn.Sequential(*list(backbone.children())[:-2])  # output channels = 512

        # Freeze backbone initially
        for param in self.feature_extractor.parameters():
            param.requires_grad = False

        # Custom CNN head operating on backbone output
        self.cnn_head = nn.Sequential(
            nn.Conv2d(512, 256, kernel_size=3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.MaxPool2d(2),

            nn.Conv2d(256, 128, kernel_size=3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.AdaptiveAvgPool2d((1,1)),
        )

        self.fc = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(64, num_classes)
        )

    def forward(self, x):
        x = self.feature_extractor(x)  # (B,512,H',W')
        x = self.cnn_head(x)           # (B,128,1,1)
        x = self.fc(x)
        return x

model = HybridResNetCNN(num_classes=len(class_names)).to(DEVICE)

# --- Optimizer & Loss
optimizer = torch.optim.Adam(model.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)
criterion = nn.CrossEntropyLoss()
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', factor=0.5, patience=2)

def accuracy_from_logits(logits, y):
    preds = torch.argmax(logits, dim=1)
    return (preds == y).float().mean().item()

# --- Training Loop (standard)
best_val_acc, history = 0.0, {"train_loss": [], "val_loss": [], "train_acc": [], "val_acc": []}
for epoch in range(1, EPOCHS+1):
    model.train()
    train_loss = train_acc = n_train = 0
    for X, y in tqdm(train_loader, desc=f"Epoch {epoch}/{EPOCHS} [Train]"):
        X, y = X.to(DEVICE), y.to(DEVICE)
        optimizer.zero_grad()
        logits = model(X)
        loss = criterion(logits, y)
        loss.backward(); optimizer.step()
        bs, acc = y.size(0), accuracy_from_logits(logits, y)
        train_loss += loss.item() * bs; train_acc += acc * bs; n_train += bs
    train_loss /= n_train; train_acc /= n_train

    model.eval(); val_loss = val_acc = n_val = 0
    with torch.no_grad():
        for X, y in val_loader:
            X, y = X.to(DEVICE), y.to(DEVICE)
            logits = model(X); loss = criterion(logits, y)
            bs, acc = y.size(0), accuracy_from_logits(logits, y)
            val_loss += loss.item() * bs; val_acc += acc * bs; n_val += bs
    val_loss /= n_val; val_acc /= n_val

    history["train_loss"].append(train_loss); history["val_loss"].append(val_loss)
    history["train_acc"].append(train_acc); history["val_acc"].append(val_acc)
    print(f"Epoch {epoch}: train_loss={train_loss:.4f} val_loss={val_loss:.4f} | train_acc={train_acc:.4f} val_acc={val_acc:.4f}")
    scheduler.step(val_acc)

    if val_acc > best_val_acc:
        best_val_acc = val_acc
        torch.save(model.state_dict(), CHECKPOINT_PATH)
        print(f"  ✔ Saved new best model (val_acc={val_acc:.4f})")

print("Best val acc:", best_val_acc)

# --- Plots
epochs = range(1, len(history["train_loss"])+1)
plt.figure(); plt.plot(epochs, history["train_loss"], label="Train Loss")
plt.plot(epochs, history["val_loss"], label="Val Loss"); plt.legend(); plt.title("Loss vs Epochs"); plt.show()
plt.figure(); plt.plot(epochs, history["train_acc"], label="Train Acc")
plt.plot(epochs, history["val_acc"], label="Val Acc"); plt.legend(); plt.title("Accuracy vs Epochs"); plt.show()

# --- Load best & Test Evaluation
model.load_state_dict(torch.load(CHECKPOINT_PATH, map_location=DEVICE))
model.eval()
all_preds, all_labels = [], []
with torch.no_grad():
    for X, y in tqdm(test_loader, desc="Testing"):
        X = X.to(DEVICE)
        preds = torch.argmax(model(X), dim=1).cpu().numpy()
        all_preds.extend(preds); all_labels.extend(y.numpy())

cm = confusion_matrix(all_labels, all_preds)
print("Classification Report:\n", classification_report(all_labels, all_preds, target_names=class_names, digits=4))
print("Confusion Matrix:\n", cm)

def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion Matrix'):
    if normalize: cm = cm.astype('float') / cm.sum(axis=1, keepdims=True).clip(min=1)
    plt.figure(); plt.imshow(cm, interpolation='nearest'); plt.title(title); plt.colorbar()
    tick_marks = np.arange(len(classes)); plt.xticks(tick_marks, classes, rotation=45, ha='right'); plt.yticks(tick_marks, classes)
    fmt = '.2f' if normalize else 'd'; thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt), ha="center", color="white" if cm[i,j] > thresh else "black")
    plt.ylabel('True label'); plt.xlabel('Predicted label'); plt.tight_layout(); plt.show()

plot_confusion_matrix(cm, class_names, normalize=False, title='Confusion Matrix (Counts)')
plot_confusion_matrix(cm, class_names, normalize=True, title='Confusion Matrix (Normalized)')

# --- Grad-CAM Implementation  ---
class GradCAM:
    def __init__(self, model, target_layer):
        """
        model: full model
        target_layer: nn.Module (conv layer) to hook
        """
        self.model = model
        self.target_layer = target_layer
        self.activations = None
        self.gradients = None
        # register hooks
        # forward hook
        def forward_hook(module, input, output):
            # keep activations as tensor on same device
            self.activations = output
        self._fhandle = target_layer.register_forward_hook(forward_hook)
        # full backward hook (future-safe)
        def backward_hook(module, grad_input, grad_output):
            # grad_output[0] is gradients w.r.t. output of the layer
            self.gradients = grad_output[0]
        # register_full_backward_hook if available, else fallback
        if hasattr(target_layer, "register_full_backward_hook"):
            self._bhandle = target_layer.register_full_backward_hook(backward_hook)
        else:
            # older PyTorch (deprecated warn), use register_backward_hook
            self._bhandle = target_layer.register_backward_hook(backward_hook)

    def generate(self, x, class_idx=None):
        """
        x: single-image tensor (1, C, H, W) on same device as model
        class_idx: if None, takes predicted class for the single image
        returns: cam as numpy 2D array (H_feat, W_feat) normalized to [0,1]
        """
        assert x.dim() == 4 and x.size(0) == 1, "input must be a single image tensor with batch size 1"
        # zero grads
        self.model.zero_grad()
        # forward
        logits = self.model(x)  # (1, num_classes)
        # choose class index
        if class_idx is None:
            class_idx = torch.argmax(logits, dim=1)[0].item()
        # create scalar loss for that sample/class
        score = logits[0, class_idx]
        # backward
        score.backward(retain_graph=True)
        # get activations and gradients (they are on device)
        activations = self.activations.detach()   # shape (1, C, H, W)
        gradients = self.gradients.detach()       # shape (1, C, H, W)
        # global-average-pool the gradients spatially -> channel weights
        weights = gradients.mean(dim=(2,3), keepdim=True)  # shape (1, C, 1, 1)
        # weighted combination of activations
        cam = (weights * activations).sum(dim=1).squeeze(0)  # shape (H, W)
        cam = F.relu(cam)
        # normalize to [0,1]
        cam -= cam.min()
        cam /= (cam.max() + 1e-8)
        cam_np = cam.cpu().numpy()
        return cam_np

# Create GradCAM hooking the first conv in cnn_head (Conv2d(512->256))
target_layer = model.cnn_head[0]
gradcam = GradCAM(model, target_layer)

# --- helper to unnormalize image and overlay heatmap
# --- Function to unnormalize and convert tensor to image ---
def unnormalize_tensor(tensor):
    mean = np.array([0.485, 0.456, 0.406])
    std  = np.array([0.229, 0.224, 0.225])
    img = tensor.permute(1,2,0).cpu().numpy()
    img = std * img + mean
    return np.clip(img, 0, 1)

# --- Function to overlay heatmap ---
def overlay_heatmap(img_tensor, cam, alpha=0.4):
    img = unnormalize_tensor(img_tensor)  # HxWx3
    cam_resized = cv2.resize(cam, (img.shape[1], img.shape[0]))

    cmap = plt.get_cmap("jet")  # use matplotlib colormap safely
    heatmap = cmap(cam_resized)[:,:,:3]   # HxWx3

    overlay = (1 - alpha) * img + alpha * heatmap
    return np.clip(overlay, 0, 1)


# --- Run Grad-CAM on up to 4 images from the test_loader
model.eval()
batch = next(iter(test_loader))  # get first batch
X_batch, y_batch = batch
X_batch = X_batch.to(DEVICE); y_batch = y_batch.to(DEVICE)

n_show = min(4, X_batch.size(0))
fig, axes = plt.subplots(2, n_show, figsize=(4*n_show, 8))
for j in range(n_show):
    img = X_batch[j].unsqueeze(0)   # (1,C,H,W)
    label = y_batch[j].item()
    cam = gradcam.generate(img)     # cam on feature map spatial size
    overlay = overlay_heatmap(img[0], cam)
    # original (unnormalized) image for display
    orig = unnormalize_tensor(img[0])
    axes[0, j].imshow(orig)
    axes[0, j].set_title(f"Original: {class_names[label]}")
    axes[0, j].axis("off")
    axes[1, j].imshow(overlay)
    axes[1, j].set_title("Grad-CAM")
    axes[1, j].axis("off")

plt.tight_layout(); plt.show()

# --- Feature Embedding (t-SNE & UMAP) ---
features, labels = [], []
model.eval()
with torch.no_grad():
    for X, y in tqdm(test_loader, desc="Extracting features"):
        X = X.to(DEVICE)
        feats = model.feature_extractor(X)   # (B,512,H',W')
        # global average pool to (B,512)
        feats = F.adaptive_avg_pool2d(feats, (1,1)).flatten(1)
        features.append(feats.cpu().numpy())
        labels.extend(y.numpy())

features = np.concatenate(features, axis=0)
labels = np.array(labels)

# --- t-SNE
# --- t-SNE
tsne = TSNE(n_components=2, random_state=42, perplexity=5)  # FIXED
features_tsne = tsne.fit_transform(features)

plt.figure(figsize=(8,6))
for i, c in enumerate(class_names):
    plt.scatter(features_tsne[labels==i,0], features_tsne[labels==i,1], label=c, alpha=0.7)
plt.legend(); plt.title("t-SNE of Features"); plt.show()

# --- UMAP
reducer = umap.UMAP(random_state=42)
features_umap = reducer.fit_transform(features)

plt.figure(figsize=(8,6))
for i, c in enumerate(class_names):
    plt.scatter(features_umap[labels==i,0], features_umap[labels==i,1], label=c, alpha=0.7)
plt.legend(); plt.title("UMAP of Features"); plt.show()
```
